```python
import geopandas as gpd
import pandas as pd
import funcs as tourism
```

### Find Boundary Data and EU Tourism Data and Merges them into GeoData

```python
tourism_data = tourism.call2016()
FILENAME = 'NUTS_Level_2_(January_2018)_Boundaries.geojson'
gdf = gpd.read_file(FILENAME)
df_geodata = tourism_data.merge(gdf)
#df_geodata.head()
```

### Plots GeoData in Accordance to the Investigation Variable

```python
investigation_var = 'Establishments_Percentage change on previous period'

gdf = gpd.GeoDataFrame(
    df_geodata, geometry= 'geometry')
gdf.plot(investigation_var)
```
